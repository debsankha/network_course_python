### Exercise (Python syntax)

**E1.1** Write a function that accepts a list as argument, adds up the numbers and returns the sum. For example:

```python
 In [1] mysum([1,4,5])
Out [1] 10
    ```

````python
def mysum(l):
    total=0
    for item in l:
        total=total+item
    return total

mysum([1,4,5])
````

####But there is a built -in function for that:

````python
sum([1,4,5])
````

**E1.2** Create a list of tuples:
```
tuplist=[('a',1),('b',2),('c',3)]
```

Now create an empty dictionary, insert the elements of `tuplist` as key->value pairs. e.g.:
```python
 In[2] print d['a']
Out[2] 1
```

````python
tuplist=[('a',1),('b',2),('c',3)]
d={}
for keyval in tuplist:
    key=keyval[0]
    val=keyval[1]
    d[key]=val

print d
````

#### Again, there's  a much simpler way of achieving this:

````python
print dict([('a',1),('b',2),('c',3)])
````
